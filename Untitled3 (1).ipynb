## **TITLE: HEAT EQUATION IMPLEMENTATION USING PYTHON**


### **1. INTRODUCTION**

This report documents the implementation of a solver for the one-dimensional heat equation using Python. The heat equation is a fundamental partial differential equation that describes how temperature changes over time in a given medium. Solving this equation numerically allows us to simulate temperature distributions in various physical systems, making it a crucial tool in fields such as physics, engineering, and materials science.

### **2. PROBLEM STATEMENT**

The objective of this implementation is to simulate the temperature distribution along a one-dimensional rod over a specified time period, governed by the heat equation. The rod has fixed boundaries, and the initial temperature distribution is provided. The solver aims to calculate the temperature at different spatial points and time steps to visualize how heat propagates through the rod over time.

### **3. METHODOLOGY**

* **Finite Difference Method:** The solver employs the explicit finite difference method to approximate the spatial and temporal derivatives in the heat equation. This involves discretizing both the spatial and temporal domains and updating the temperature at each grid point based on neighboring values.
* **Initialization:** The spatial grid and temporal grid are initialized, along with the parameters such as rod length, total time, thermal diffusivity, and initial temperature distribution.
* **Boundary and Initial Conditions:** The boundaries of the rod are assumed to be insulated, and the initial temperature distribution is provided as a function of position along the rod.
* **Numerical Solution:** The solver iterates over time steps, updating the temperature at each spatial grid point based on the finite difference approximation of the heat equation. The process continues until the specified simulation time is reached.
* **Visualization:** Matplotlib is used to visualize the temperature distribution along the rod at different time steps, allowing for the observation of temperature evolution over time.


### **4. PYTHON IMPLEMENTATION**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 1.0         # Length of the rod
T = 0.5         # Total time
Nx = 100        # Number of spatial grid points
Nt = 1000       # Number of time steps
alpha = 0.01    # Thermal diffusivity

# Spatial grid
x = np.linspace(0, L, Nx)

# Time grid
t = np.linspace(0, T, Nt)

# Grid spacing
dx = x[1] - x[0]
dt = t[1] - t[0]

# Initial condition: Temperature distribution at t = 0
u_initial = np.sin(np.pi * x)

# Initialize temperature array
u = np.zeros((Nt, Nx))

# Set initial condition
u[0, :] = u_initial

# Main loop for time integration
for n in range(Nt - 1):
    for i in range(1, Nx - 1):
        u[n+1, i] = u[n, i] + alpha * dt / dx**2 * (u[n, i-1] - 2 * u[n, i] + u[n, i+1])

# Plot the temperature distribution at different time steps
plt.figure(figsize=(10, 6))
for i in range(0, Nt, Nt//10):  # Plotting every 10th time step
    plt.plot(x, u[i, :], label=f"t={t[i]:.2f}")
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.title('Temperature Distribution over Time')
plt.legend()
plt.grid(True)
plt.show()
```

### **5. RESULTS:**

* The implemented solver successfully simulates the temperature distribution along the rod over the specified time period.
* The temperature distribution evolves over time, demonstrating the diffusion of heat through the rod.
* The plotted temperature profiles provide a visual representation of how temperature changes at different positions along the rod over time.
